```python
## Write a function in Python to check whether a number is Prime or not.
```

```python
def prime(num):
    if num > 1:
        for i in range(2,num):
            if(num % i) == 0:
                print(num,'is not a prime number')
                print(i,'times',num//i,'is',num)
                break
        else:
            print(num,'is a prime number')
    else:
        print(num,'is not a prime number')
```

```python
prime(71)
```
